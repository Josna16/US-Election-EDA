## 2.Cleaning the Transcripts

```python
#first step of cleaning the transcripts
# defining function for changing all text into lower case, removing text within square-bracket, removing punctuation marks, removing words with digits in them
import re
import string

def clean_text_round1(text):
    text = text.lower()
    text = re.sub('[.*]', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round1 = lambda u: clean_text_round1(u)
```

```python
#dataframe after first step of cleaning
data_clean_joe = pd.DataFrame(data_df_joe2.Transcript.apply(round1))
data_clean_joe    
```

```python
data_clean_trump = pd.DataFrame(data_df_trump2.Transcript.apply(round1))
data_clean_trump    
```

```python
#second step of cleaning transcript
#defining function for removing new-line character and some additional punctuation marks 
def clean_text_round2(text):
    text = re.sub('\n', '', text)
    text = re.sub('[‘’“”…]', '', text)
    return text

round2 = lambda u: clean_text_round2(u)
```

```python
#dataframe after second step of cleaning
data_clean_joe = pd.DataFrame(data_clean_joe.Transcript.apply(round2))
data_clean_joe
```

```python
data_clean_trump = pd.DataFrame(data_clean_trump.Transcript.apply(round2))
data_clean_trump    
```

```python
#using sklearn to find the word count for each word in each state
from sklearn.feature_extraction.text import CountVectorizer

cv_joe = CountVectorizer(stop_words='english')
data_cv_joe = cv_joe.fit_transform(data_clean_joe.Transcript)
data_dtm_joe = pd.DataFrame(data_cv_joe.toarray(), columns=cv_joe.get_feature_names())
data_dtm_joe.index = data_clean_joe.index
data_dtm_joe
```

```python
cv_trump = CountVectorizer(stop_words='english')
data_cv_trump = cv_trump.fit_transform(data_clean_trump.Transcript)
data_dtm_trump = pd.DataFrame(data_cv_trump.toarray(), columns=cv_trump.get_feature_names())
data_dtm_trump.index = data_clean_trump.index
data_dtm_trump
```

```python
data_joe1 = data_dtm_joe.transpose()
data_joe1
```

```python
#dictionary consisting of topwords and their counts for each states in the descending order
top_dict_joe = {}

for c in data_joe1.columns:
    top = data_joe1[c].sort_values(ascending=False).head(30)
    top_dict_joe[c] = list(zip(top.index, top.values))

top_dict_joe
```

```python
#printing topwords for each state
for state, topwords in top_dict_joe.items():
    print(state)
    print(', '.join(word for word, count in topwords))
    print('---')
```

```python
data_trump1 = data_dtm_trump.transpose()
data_trump1
```

```python
top_dict_trump = {}

for c in data_trump1.columns:
    top = data_trump1[c].sort_values(ascending=False).head(30)
    top_dict_trump[c] = list(zip(top.index, top.values))

top_dict_trump
```

```python
for state, topwords in top_dict_trump.items():
    print(state)
    print(', '.join(word for word, count in topwords))
    print('---')
```

```python
#combining the topwords for all states
from collections import Counter

words_joe = []
for state in data_joe1.columns:
    top = [word for (word, count) in top_dict_joe[state]]
    for t in top:
        words_joe.append(t)

words_joe
```

```python
Counter(words_joe).most_common()
```

```python
#adding the words occuring a total of more than 6 times in all the states into the list of stopwords
add_stop_words_joe = [word for (word, count) in Counter(words_joe).most_common() if count>6]
add_stop_words_joe
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer

stop_words_joe = text.ENGLISH_STOP_WORDS.union(add_stop_words_joe)

cv_joe1= CountVectorizer(stop_words=stop_words_joe)
data_cv_joe1 = cv_joe1.fit_transform(data_clean_joe.Transcript)
data_stop_joe = pd.DataFrame(data_cv_joe1.toarray(), columns = cv_joe1.get_feature_names())
data_stop_joe.index = data_clean_joe.index
data_stop_joe
```

```python
words_trump = []
for state in data_trump1.columns:
    top = [word for (word, count) in top_dict_trump[state]]
    for t in top:
        words_trump.append(t)

words_trump
```

```python
Counter(words_trump).most_common()
```

```python
add_stop_words_trump = [word for (word,count) in Counter(words_trump).most_common() if count>7]
add_stop_words_trump
```

```python
stop_words_trump = text.ENGLISH_STOP_WORDS.union(add_stop_words_trump)

cv_trump1= CountVectorizer(stop_words=stop_words_trump)
data_cv_trump1 = cv_trump1.fit_transform(data_clean_trump.Transcript)
data_stop_trump = pd.DataFrame(data_cv_trump1.toarray(), columns = cv_trump1.get_feature_names())
data_stop_trump.index = data_clean_trump.index
data_stop_trump
```
